# **create New time Feature**

```python
df_realita["waktu"].unique()
time_list = df_realita["waktu"].unique().tolist()
```

```python
import re
from datetime import datetime, timedelta

def parse_waktu(waktu_str):
    if pd.isnull(waktu_str):
        return None

    waktu_str = waktu_str.lower().replace("diedit", "").strip()
    now = datetime.now()

    match = re.findall(r'\d+', waktu_str)

    if "jam lalu" in waktu_str and match:
        return now - timedelta(hours=int(match[0]))

    elif "hari lalu" in waktu_str and match:
        return now - timedelta(days=int(match[0]))

    elif "minggu lalu" in waktu_str and match:
        return now - timedelta(weeks=int(match[0]))

    elif "bulan lalu" in waktu_str and match:
        return now - timedelta(days=int(match[0]) * 30)

    elif "tahun lalu" in waktu_str and match:
        return now - timedelta(days=int(match[0]) * 365)

    elif "sebulan lalu" in waktu_str:
        return now - timedelta(days=30)

    elif "setahun lalu" in waktu_str:
        return now - timedelta(days=365)

    elif "seminggu lalu" in waktu_str:
        return now - timedelta(weeks=1)

    return None
parsed_time_list = [parse_waktu(t).date() if parse_waktu(t) is not None else None for t in time_list]
```

```python
for original, parsed in zip(time_list, parsed_time_list):
    print(f"{original:25} ‚Üí {parsed}")

df_realita['waktu'] = df_realita['waktu'].apply(parse_waktu)
```

# **Preprocessing Step **

# 1.Lowercase

```python
df_realita['review_lowercase'] = df_realita['review'].astype(str).str.lower()
```

```python
df_realita['review_lowercase']
```

# 2.Change emoji Into text

```python
emoji_df = pd.read_csv('/content/drive/MyDrive/Scrapping Competition /emoji_to_text.csv')
emoji_dict = dict(zip(emoji_df['emoji'], emoji_df['makna']))
def replace_emojis(text):
    for emoji, makna in emoji_dict.items():
        if emoji in text:
            text = text.replace(emoji, f" {makna} ")
    return text
df_realita['review_emoji'] = df_realita['review_lowercase'].apply(replace_emojis)
```

3.**Puctuation**

```python
import re

def clean_text(text):
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_realita['review_punc'] = df_realita['review_emoji'].apply(clean_text)
```

# **4.laguage Detection and we decide to tranlate english into indonesia**

```python
from langdetect import detect
from langdetect.lang_detect_exception import LangDetectException

def detect_language(text):
    try:
        return detect(text)
    except LangDetectException:
        return 'unknown'

df_realita['language'] = df_realita['review_punc'].apply(lambda x: detect_language(str(x)))
print(df_realita['language'].value_counts())
```

```python
from googletrans import Translator

translator = Translator()
df_english = df_realita[df_realita['language'] == 'en'].copy()

translated = []
for review in df_english['review_punc']:
    try:
        result = translator.translate(review, src='en', dest='id')
        translated.append(result.text)
    except Exception as e:
        translated.append("TRANSLATION_FAILED")


df_english['review_trans'] = translated
df_realita.loc[df_realita['language'] == 'en', 'review_trans'] = df_english['review_trans'].values
df_realita['review_trans'] = df_realita['review_trans'].fillna(df_realita['review_punc'])

print(df_realita[['review_punc', 'language', 'review_trans']].head())
```

```python
import nltk
from nltk.corpus import stopwords

# Download stopwords bahasa Indonesia
nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))

# Hapus stopwords dari token list
df_realita['review_stopwords'] = df_realita['review_tokens'].apply(
    lambda tokens: [word for word in tokens if word not in stop_words]
)
```

# **Slang words changer**

```python
slang_dict = {}
with open("/content/drive/MyDrive/Scrapping Competition /combined_slang_words.txt", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        if len(parts) == 2:
            slang, formal = parts
            slang_dict[slang.strip()] = formal.strip()

# 4. Replace slang words (hasilnya tetap list)
def replace_slang(tokens):
    return [slang_dict.get(w, w) for w in tokens]

df_realita['review_slang'] = df_realita['review_stopwords'].apply(replace_slang)
```
