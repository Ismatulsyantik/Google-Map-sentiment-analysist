```python
import pandas as pd
import torch, torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 1. Load data
df_realita = pd.read_csv("/content/realita_cleand_final.csv")
docs = df_realita['review_slang'].dropna().tolist()

# 2. Load Sentiment Model IndoBERT
model_id = "mdhugol/indonesia-bert-sentiment-classification"
tokenizer = AutoTokenizer.from_pretrained(model_id)
sent_model = AutoModelForSequenceClassification.from_pretrained(model_id)

labels = ["negative", "neutral", "positive"]

# 3. Fungsi batch sentiment
def get_sentiment_batch(texts, batch_size=16):
    results = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        inputs = tokenizer(batch, return_tensors="pt", truncation=True,
                           padding=True, max_length=512)
        with torch.no_grad():
            outputs = sent_model(**inputs)
            probs = F.softmax(outputs.logits, dim=-1).tolist()
        # simpan hasil
        for p in probs:
            results.append(dict(zip(labels, p)))
    return results

# 4. Hitung sentimen semua review dalam batch
sentiments = get_sentiment_batch(docs, batch_size=32)

# 5. Masukkan ke DataFrame
df = pd.DataFrame({"review": docs, "sentiment": sentiments})
sent_df = pd.json_normalize(df["sentiment"])
df = pd.concat([df, sent_df], axis=1)

# 6. Agregasi rata-rata
overall_sentiment = df[["negative","neutral","positive"]].mean()
print("Overall Sentiment Score:")
print(overall_sentiment)

# Contoh hasil per review
print(df.head())
```

```python
df.to_csv('realita_sentiment.csv', index=False)
```
